## MITRE

#### Here in this notebook i have made data processing the coolected was in json processed them into document by retriving description, mitigation stratergy cwe cve etc

```python
import json
import os
```

```python
inputf="../datasets/mitre/course-of-action"
outputf="../prepro_doc/mitre_docs"
```

```python
os.makedirs(outputf,exist_ok=True)
count=0
```

```python
count=0

for file in os.listdir(inputf):
    if file.endswith(".json"):
        path=os.path.join(inputf, file)
        with open(path,"r",encoding="utf-8") as f:
            data=json.load(f)
        obj=data["objects"][0]
        if obj.get("x_mitre_deprecated")==True:
            continue 
        name=obj.get("name","")
        desc=obj.get("description","")
        text=f"""
MITRE Mitigation

Name: {name}

Description:
{desc}
"""
        output_path=os.path.join(outputf, f"mitre_{count}.txt")
        with open(output_path,"w",encoding="utf-8") as f:
            f.write(text)
        count=count+1

print("all good", count)
```

### CVE

```python
inputfcve="../datasets/cve/nvdcve-2.0-2023.json"
outputfcve="../prepro_doc/cve_docs"

os.makedirs(outputf, exist_ok=True)
```

```python
count=0 
with open(inputf,"r",encoding="utf-8") as f:
    data=json.load(f)

for item in data["vulnerabilities"][:100]:
    cve_id=item["cve"]["id"]
    desc=item["cve"]["descriptions"][0]["value"]
    store=f"""
CVE Vulnerability

ID: {cve_id}

Description:
{desc}
"""
    output_path=os.path.join(outputf,f"cve_{count}.txt")

    with open(output_path,"w",encoding="utf-8") as f:
        f.write(store)

    count=count+1

print("all good", count)
```

### CWE

```python
import pandas as pd
import re
```

```python
inputfcwe="../datasets/cwe/699.csv"
outputfcwe="../prepro_doc/cwe_docs"

os.makedirs(outputfcwe,exist_ok=True)
```

```python
df=pd.read_csv(inputfcwe)
print(df.columns)
```

```python
count=0

for index, row in df.iterrows():
    cwe_id=row["CWE-ID"]
    name=row["Name"]
    desc=row["Description"]
    mitigation_text=str(row["Potential Mitigations"])
    mitigations=re.findall(r"DESCRIPTION:(.*?)(?:::|$)",mitigation_text)
    clean_mitigation="\n".join(mitigations)
    if pd.isna(desc):
        continue
    doc=f"""
CWE Weakness

ID: CWE-{cwe_id}

Name: {name}

Description:
{desc}

Mitigation:
{clean_mitigation}
"""
    outpath=os.path.join(outputfcwe,f"cwe_{count}.txt")

    with open(outpath,"w",encoding="utf-8") as f:
        f.write(doc)
    count += 1
print("CWE documents created:", count)
```
