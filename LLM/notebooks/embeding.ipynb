```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
from sentence_transformers import SentenceTransformer
import os

```

```python
basef="../prepro_doc"
docs=[]
```

```python
for root,dirs,files in os.walk(basef):
    for file in files:
        if file.endswith(".txt"):
            path=os.path.join(root,file)
            with open(path,"r",encoding="utf-8") as f:
                text=f.read()
            docs.append(Document(page_content=text))
print("done",len(docs))
```

```python
docs
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embed=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
vector_db=FAISS.from_documents(docs,embed)
```

```python
vector_db.save_local("../vector_db")
```

```python
vector_db=FAISS.load_local("../vector_db",embed,allow_dangerous_deserialization=True)
```

```python
question="How does previlage escalation work?"
res=vector_db.similarity_search(question,k=3)
i=0
for r in res:
    print("Document ",i+1)
    print(r.page_content)
    i+=1
```
